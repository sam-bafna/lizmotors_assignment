```python
from transformers import BertTokenizer, BertModel
# Your text
text = [
 'The stale smell of old beer lingers.',
 'It takes heat to bring out the odor.',
 'A cold dip restores health and zest.',
 'A salt pickle tastes fine with ham.',
 'Tacos al pastor are my favorite.',
 'A zestful food is the hot cross bun.'
]
# Initialize tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
# Tokenize text
inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
# Forward pass, get pooled_output
outputs = model(**inputs)
pooled_output = outputs.pooler_output
# pooled_output now contains the embeddings for your sentences
print(pooled_output.shape)
```

```python
from gtts import gTTS

text_to_say = """
The stale smell of old beer lingers.
It takes heat to bring out the odor.
A cold dip restores health and zest.
A salt pickle tastes fine with ham.
Tacos al pastor are my favorite.
A zestful food is the hot cross bun.
A good beer is the beer of the week.
A good beer is the beer of the month.
A good beer is the beer of the year.
A good beer is
"""

language = "en"

gets_object = gTTS(text=text_to_say,
                  lang=language,
                  slow=False)

gets_object.save("/content/harvard.wav")
```
